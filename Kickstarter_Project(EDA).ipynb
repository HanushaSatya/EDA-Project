```python
# Step 12: Average Backers per Category
data['Backers'] = pd.to_numeric(data['Backers'], errors='coerce')

avg_backers = data.groupby('Category')['Backers'].mean().sort_values(ascending=False)

avg_backers.plot(kind='bar', color='salmon')
plt.title('Average Number of Backers per Category')
plt.xlabel('Category')
plt.ylabel('Average Backers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 12: Success Rate by Goal Range
data['goal_range'] = pd.cut(data['Goal'], bins=[0, 1000, 5000, 10000, 20000, 50000, 100000, 500000, 1e6, 1e7],
                          labels=['<1k','1kâ€“5k','5kâ€“10k','10kâ€“20k','20kâ€“50k','50kâ€“100k','100kâ€“500k','500kâ€“1M','1M+'])

goal_success_rate = (
    data[data['State'] == 'Successful'].groupby('goal_range').size()
    / data.groupby('goal_range').size()
)

goal_success_rate.plot(kind='bar', color='teal')
plt.title('Success Rate by Goal Range')
plt.xlabel('Goal Range (USD)')
plt.ylabel('Success Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  Step 13: Number of Backers Distribution (Log Scale)
plt.figure(figsize=(12,7))
sns.histplot(data['Backers'], bins=100, color='orchid', log_scale=True)
plt.title('Distribution of Number of Backers (Log Scale)')
plt.xlabel('Backers')
plt.ylabel('Number of Projects')
plt.tight_layout()
plt.show()
```

```python
# Step 14: Average Pledge per Backer (Only Successful Projects)
df_success = data[(data['State'] == 'Successful') & (data['Backers'] > 0)]
df_success['avg_pledge_per_backer'] = df_success['Pledged'] / df_success['Backers']

sns.histplot(df_success['avg_pledge_per_backer'], bins=100, color='coral')
plt.title('Average Pledge Per Backer (Successful Projects)')
plt.xlabel('Avg Pledge ($)')
plt.ylabel('Number of Projects')
plt.tight_layout()
plt.show()
```

```python
# Calculate success rates per category
category_success = df[df['State'] == 'Successful']['Category'].value_counts()
category_total = df['Category'].value_counts()
success_rate = (category_success / category_total).sort_values(ascending=False)

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=success_rate.index, y=success_rate.values, palette='viridis')
plt.xticks(rotation=45)
plt.title('Success Rate by Project Category')
plt.ylabel('Success Rate')
plt.xlabel('Project Category')
plt.tight_layout()
plt.show()
```

```python
df['goal_bin'] = pd.cut(df['Goal'], bins=[0, 1000, 5000, 10000, 20000, 50000, 100000, float('inf')],
                        labels=['<1k', '1k-5k', '5k-10k', '10k-20k', '20k-50k', '50k-100k', '>100k'])

# Calculate success rates per goal bin
goal_success = df[df['State'] == 'Successful']['goal_bin'].value_counts()
goal_total = df['goal_bin'].value_counts()
goal_success_rate = (goal_success / goal_total).sort_index()

# Plotting
plt.figure(figsize=(10, 5))
sns.lineplot(x=goal_success_rate.index, y=goal_success_rate.values, marker='o')
plt.title('Success Rate by Funding Goal')
plt.ylabel('Success Rate')
plt.xlabel('Funding Goal Range')
plt.tight_layout()
plt.show()
```

```python
# Calculate number of successful campaigns per country
country_success = df[df['State'] == 'Successful']['Country'].value_counts().head(10)

# Plotting
plt.figure(figsize=(12, 5))
sns.barplot(x=country_success.index, y=country_success.values, palette='coolwarm')
plt.title('Top 10 Countries by Number of Successful Campaigns')
plt.ylabel('Number of Successful Campaigns')
plt.xlabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Drop rows with NaT in 'launched' column
df = df.dropna(subset=['Launched'])

# Extract launch month
df['launch_month'] = df['Launched'].dt.month_name()
df.sort_values(by='launch_month',ascending=True,inplace=True)

# Calculate total and successful campaigns per month
monthly_total = df['launch_month'].value_counts().sort_index()
monthly_success = df[df['State'] == 'Successful']['launch_month'].value_counts().sort_index()
monthly_success_rate = (monthly_success / monthly_total).sort_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_success_rate.index, y=monthly_success_rate.values, marker='o', color='Skyblue')
plt.title('Monthly Launch Success Rates')
plt.ylabel('Success Rate')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
